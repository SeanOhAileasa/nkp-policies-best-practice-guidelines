###### 2. Privileged User Agreement
***

After completing this topic, you should be able to: ``identify privileged user agreements``

In this presentation, we'll take a look at the Privileged User Agreement, which is typically a document that defines what privileged access means.

And in almost every computing environment, there is at least one user who has elevated powers within a specific computer system or for the entire network.

And that quite simply means they have greater power than a regular system user, they can do things that standard users cannot do. 

So the privileged users may have access to systems that can affect those individual computing systems or overall network communication, the user accounts that are in use and, of course, the files and folders. 

So, quite simply, they oversee the day-to-day operations of the IT environment. 

So then privileged access is typically granted to system administrators or networks administrators, and you might hear of those two terms used fairly interchangeably. But, I guess, at the most official levels, system administrators are responsible for specific systems, such as messaging servers or database servers. 

Whereas network administrators would be more so concerned with the routers, the switches, the utilization of bandwidth, firewalls, the security, demilitarized zones, things along those lines. But, again, in day-to-day conversation you'll hear those two terms used fairly interchangeably. 

But you might also see select staff that are required to perform special duties over a computing system or a network that might be dedicated to specific departments. For example, there could be an IT person specifically for accounting or human resources. Or, again, there might just be a dedicated system within the organization, such as a healthcare system. So you might have specific people that have some elevated capabilities within those departments or for those systems. 

As for the expectations, users with privileged access must respect the rights of the system users, the integrity of the system itself, and any peripheral or related physical resources. And, of course, they also have to comply with policies of the organization, laws, and any other regulations that might be imposed by any kind of commission or just the business itself. 

Other considerations, the user who has the privileged access has to be authorized. In other words, whoever it is going to be, they typically have to sign a Privileged User Agreement. And in some cases, they might even need to be security cleared, particularly in the military, for example, there are various levels of security clearance. 

And the job duties, effectively, should be minimized when it comes to the privileged access, meaning that you should only use those privileges when they are necessary to perform the required task. Or, you look at from the other perspective, if any particular duty can be performed in another manner, then avoid using the privileged access. So if, for example, you are just logging on to your workstation to check your email, you should not log on as an administrator. You should create a separate account for yourself that is just a regular user, and for your standard day-to-day tasks you should log in as the regular user. Then when you need to perform those elevated tasks, then certainly log in as the administrator or invoke the administrative credentials. But, generally, it means, do not log in as an administrator all the time, every day. Because if your account should be compromised, for example, you just go up to get a coffee and you leave your system logged in, then anyone else could sit down at that station and perform tasks as an administrator. Also, if any kind of malware should find its way onto your system and you are logged in as an administrator, then the malware typically runs as an administrator as well. So again, whenever possible, log in as a regular user. 

Some other considerations, with respect to the duties that are generally involved with those elevated capabilities, include running security software, monitoring systems for reliability and security, installing system software. Moving or archiving files and folders, certainly, implementing repairs, fixing files or file processes, stopping or starting services. Killing processes that seem to be consuming too many resources, what they refer to often as runaway processes, or those that might seem to be invasive or intrusive processes. 

And when it comes to some of those duties, again, it often involves the ability to grant, change, or deny access to any given resource to other users. So, of course, you are in charge of saying you may access this, you can't access this and/or, of course, making changes. You might also be able to grant your privileges, or at least a certain level of elevated privileges, to other people. So it might be at your discretion to say, okay, this person is going to be promoted, if you will, to a junior administrator. Therefore, I am going to grant them the following privileges. And you would ultimately be responsible then for whatever that person does because you gave them the ability to do so. 

Some other examples might include disabling an account that was abusing a system, such as sending harassing emails, or, of course, blocking network intrusion attempts. 

But there can be a lot of responsibility that falls on you as the privileged user, so you have to be willing to undertake those responsibilities and, of course, willing to face the consequences in the event that something does go wrong. 

So again, it's just something that should be clearly outlined with respect to what the responsibilities are. 

And, of course, it's something that should be known very well by both sides, you as the privileged user and the powers that be, if you will, so that everybody understands what is expected. 

And sometimes that involves you, as the administrator, informing them that this should not be done this way or this should be changed. Because sometimes, you might have a higher level of technical knowledge than those who are hiring you. And that's fine, but as long as everybody understands how things should be implemented, then these are the things that you want to include and are outlined in the Privileged User Agreement.

###### IT Offboarding Checklist
***

[IT Offboarding Checklist. ] So again this is something that should be taken care of as soon as possible.

###### Software Licence
***

[Software License. ] Now this is a document providing legally binding guidelines that outline the use and the distribution of software. 

And licensing provides organizations with this documentation to use one or more copies of the software while ensuring that no copyrights are violated.

It also defines responsibilities of parties purchasing the software product and outlines any restrictions on how it can be used. 

Now for the most part those restrictions deal with redistributing the software. You are not allowed to take someone else's product, repackage it and sell it for your own gain.

###### Software Licensing Terms
***

[Software Licensing Terms. ] Now most licensing agreements will include terms of the license, including statements such as a fair use. [of software. ] And this refers to, quite literally, how it is intended to be used and what is its purpose. 

Now something that you might need to consider here is what's known as reverse engineering. [the software. ] That in and of itself is perfectly legal. ###### Types of Software Licenses
***

[Types of Software Licenses. ] Now in terms of the types of software licenses, there are a few different categories. 

The first being proprietary, which means that the product is copyrighted which bears limits against its use, its distribution, and modification. And it remains the property of the software owner or the creator. So organizations acquire and use the software under predefined conditions imposed by the owner. In other words, they license a copy. And once you obtain it you are of course allowed to use it for the purpose that was intended by the developer. And this is primarily what you see in most corporate environments most of the time. 

But there is also, of course, free software. And this is any kind of application that can be freely acquired, used, modified, and even redistributed. But the only restriction is that redistribution of the software must follow the original terms. For example, again, you are not allowed to take free software, repackage it and then sell it for personal gain. It has to effectively remain as free software. 

And finally, open source, which is similar to free software in that the source code of a program is made available to the public free of charge. And the source code can be used or modified as per the needs of the developer who is actually implementing it. But whereas free software is typically just created by a single entity or a single person, open source code is often developed as part of a public collaboration. There is a community, if you will, wherein everybody can have a say. But the model itself is that the software must be redistributed without any restrictions. The source code must remain freely available. And improved versions of the software can carry a different name if outlined and documented in the licensing. So there are a few more restrictions or controls, if you will, on open source software, but it is still free. That's certainly attractive to a lot of organizations. If you can find an open source application that suits your needs, then clearly you can save some money. But one of the drawbacks with both free software and open source is that there really is no support. The vendor does not provide any kind of support. In a lot of cases, there really is no specific vendor. It's not being sold. So you can certainly find a lot of documentation and a lot of forums on the Internet that can help you out. But when you purchase proprietary software the license also typically defines what kind of support is available. 

So ultimately it's your decision, but there are certainly pros and cons of both. So I would put a little bit of extra research into any given type of application before you actually implement it to find out what are some of the advantages and/or disadvantages of any given type of license.

###### 6. International Export Controls
***

After completing this topic, you should be able to: ``describe international export controls``

In this presentation, we'll take a look at some import and export controls with respect to technology. [Import/Export Controls. ]

And this deals with the fact that there are federal US regulations in place that do limit the export of computers. And perhaps more specifically, encryption technology. 

Now this actually dates back to the Second World War and the German implementation of the Enigma machine. Because of course this represented a method of communication that was unbreakable to the United States and the Allied forces. And while it was eventually broken, they did not want their own encryption technology to find its way out of the country, so that it might possibly be used against them. 

So those regulations actually persisted for quite some time, well into the 1990s in fact. 

But while they were previously strict, they are certainly a lot more relaxed now to encourage open commerce. 

And this is due to the fact that encryption technology has essentially become the cornerstone of e-commerce. 

So again, much more relaxed these days than before. That said, there are still some restrictions. 

So high-performance computing systems these days can be exported to almost every country. But pre-approval by the Department of Commerce is required for Tier 3 countries, and it's entirely still prohibited to Tier 4 countries. 

Now the tiers at a very general level, refer to what kind of adherence to the law is observed in those countries. And the higher your tier, in essence, the more lawless the society. A lot of them deal with human trafficking in fact. So again the idea is we don't want to expose encryption technology and a method for them to secure their communications to use them for criminal activity. 

So again, the Tier 1 and the Tier 2 observe the law much more closely. So, those ones typically do not have any issues with respect to exporting the technology. 

So, if you do have this encryption technology that needs to be exported, [freely ] it can be done, provided you submit that technology to the Department of Commerce for review. And that again goes up to the Tier 3 countries. So if you are an exporter, and you need to export this technology then you do need this review. Otherwise, it would be considered smuggling. So, the review in and of itself should take no longer than 30 days. And you pretty much have to wait for the review to come back, and from that point, you certainly have to adhere to whatever the review says. If they say yes, it's okay, then you can certainly go ahead and export it. If they say no, then clearly you cannot. 

So again, it deals with where it is going, for the most part, and/or where it's coming from, so you always have to be mindful of that. But again, for the most part these days, if you're dealing with typically first world nations, it should not be a problem when it comes to importing and/or exporting this type of technology.

###### 7. Data Loss Prevention (DLP)
***

After completing this topic, you should be able to: ``describe data loss prevention techniques``

Now in this presentation, we'll take a look at data loss prevention, or as it is commonly referred to as DLP. 

But to quickly clarify, data loss in this context does not refer to the fact that you might have lost a file because you accidentally deleted it or because a hard drive crashed. But rather, it refers to policies and/or restrictions that have been implemented to control how valuable data is used. 

So things such as intellectual property, confidential corporate data, or personally identifiable information needs to be kept within the organization. 

So DLP is, in essence, an approach that is designed to prevent or at least minimize data leakage. So it's this data disclosure outside of the organization that is the essence of data loss. It's not lost because it's gone. It's lost because it found its way into the wrong hands. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Data Loss Prevention (DLP) Risks
***

So some of the risks associated with DLP include malware. And this would probably not really be something like a virus that is designed to destroy the data, but rather, some kind of spyware that is designed to collect it and report it back to whoever is trying to obtain it. 

Social media, people might just post things, maybe even inadvertently, that they shouldn't have, revealing too much information. 

Certainly social engineering and/or phishing scams, people or entities posing as legitimate IT staff, for example, to trick you into revealing your credentials.
